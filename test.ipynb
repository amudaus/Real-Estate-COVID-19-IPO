```python
import pandas as pd
import numpy as np
```

```python
xls = ['BOXRSA', "CHXRSA", "DAXRSA", "DEXRSA", "LXXRSA", "NYXRSA", "PHXRSA", "SEXRSA", "SFXRSA", "EMPLOY", "UNRATE", "GDPC1", "CPIAUCSL"]
# Merge all the data into one dataframe and save it as a csv file keep dates as index, and columns as the names of the files
df = pd.DataFrame()
for x in xls:
    data = pd.read_excel(f"{x}.xls", index_col=0, skiprows=10)
    df = pd.concat([df, data], axis=1)
    if x == "EMPLOY":
        df.rename(columns={"LREM64TTUSM156S": "emp_rate"}, inplace=True)
    elif x == "UNRATE":
        df.rename(columns={"UNRATE": "unemp_rate"}, inplace=True)
    elif x == "GDPC1":
        df.rename(columns={"GDPC1": "gdp"}, inplace=True)
    elif x == "CPIAUCSL":
        df.rename(columns={"CPIAUCSL": "inflation"}, inplace=True)
    else:
        df.rename(columns={f"{x}": f"{x.lower()}"}, inplace=True)
# Interpolate missing values
df.interpolate(inplace=True)
# Drop rows with missing values
df.reset_index(inplace=True)
# Replace nan with median
df = df.fillna(df.median())
df.to_csv("data.csv")
```
